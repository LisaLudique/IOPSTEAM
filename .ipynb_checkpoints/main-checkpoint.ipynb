```python
t = ["TERRORISM"]
for d in data["TERRORISM"]:
    if d == "YES":
        t.append(1)
    elif d == "NO":
        t.append(0)
    else:
        t.append(-1)
```

```python
delete = ["TERRORISM"]
with open("data.csv") as infile, open("data1.csv", "wb") as outfile:
    r = csv.DictReader(infile)
    firstrow = next(r)
    fields = r.fieldnames
    w = csv.DictWriter(outfile, [field for field in fields if not field in delete],
                      extrasaction="ignore")
    w.writeheader()
    w.writerow(firstrow)
    for row in r:
        w.writerow(row)
```

```python
from shutil import move
import csv
from tempfile import NamedTemporaryFile
with open('data1.csv') as csvfile, NamedTemporaryFile("w", dir='.', delete=False) as temp:
    r = csv.reader(csvfile)
    wr = csv.writer(temp, delimiter=',')
    for row, new in zip(r, t):
        wr.writerow(row + [new])
move(temp.name, "data.csv")
```
